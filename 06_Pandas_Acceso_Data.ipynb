We learned how to get rows using `iloc[]` before. But we can do a lot more with it. We can actually get rows *and* columns.

## iloc

### Notas

* `df.iloc[]`: Selecciona filas y columnas por posición.

### Ejemplos
Usando `iloc` vamos a:

1. Obtener la primera fila (índice 0).
2. Obtener el elemento `job_skills` de la tercera fila (índice 2).
3. Obtener `job_skills` (índice 15) y `job_type_skills` (índice 16) para la tercera (índice 2) y cuarta (índice 3) filas.
4. Obtener las primeras 12 filas del DataFrame.
5. Obtener las primeras cinco columnas del DataFrame y todas las filas.

Para esto necesitaremos conocer los números de índice de nuestro DataFrame.

![image](images/iloc_visual_1.png)


1. Obtener la primera fila.

```python
df.iloc[0]
```

2. Obtener el elemento `job_skills` de la tercera fila (índice 2).

```python
df.iloc[2][15]
```

##### Nota: Usa `df.iloc[0, 15]` en lugar de `df.iloc[0][15]` para asegurar la compatibilidad futura con pandas.

El uso de indexación encadenada como `df.iloc[0][15]` está siendo obsoleto en pandas, ya que puede llevar a un comportamiento ambiguo entre el acceso basado en posición y el basado en etiquetas en futuras versiones. Al usar `df.iloc[0, 15]`, especificas directamente la posición de los datos que deseas acceder, lo cual es más claro y evita posibles errores futuros cuando pandas cambie cómo se interpretan las claves enteras en la indexación de series.

Entonces, en su lugar deberíamos escribir:


```python
df.iloc[2,15]
```

3. Obtener `job_skills` (índice 15) y `job_type_skills` (índice 16) para la tercera (índice 2) y cuarta (índice 3) filas.
    * Para obtener la tercera (índice 2) y cuarta (índice 3) filas: `[2,3]`
    * Para obtener `job_skills` y `job_type_skills` que son los índices 15 y 16 respectivamente: `[15,16]`
    * Luego, coloca esos dos en una lista para obtener todo entre ellos: `df.iloc[[2,3],[15,16]]`

![image2](images/iloc_visual_2_a.png)


```python
df.iloc[[2,3],[15,16]]
```

## loc

### Notas

* `df.loc[]`: Selecciona filas y columnas por posición o etiqueta.
* Similar a `df.iloc[]` excepto que podemos usar etiquetas en su lugar.

### Ejemplo

Vamos a obtener las mismas columnas que antes pero con `loc`. Esto utiliza etiquetas de columnas y filas. Estas son:

1. Obtener la primera fila (índice 0).
2. Obtener las primeras 10 filas de `job_skills` y `job_type_skills`.
3. Obtener las primeras 5 columnas y las filas 10-20.
4. Obtener las primeras 12 filas del DataFrame.
5. Obtener las primeras 5 columnas del DataFrame y todas las filas.

1. Obtener la primera fila. Esto permanece igual porque la fila no tiene una etiqueta.

```python
df.loc[0]
```

2. Obtener las primeras 10 filas de `job_skills` y `job_type_skills`.

```python
df.loc[:9,['job_skills','job_type_skills']]
```

```python
df.loc[:9][['job_skills','job_type_skills']]
```

3. Obtener las primeras 5 columnas y las filas 10-20.

```python
df.loc[10:20,'job_title_short':'job_work_from_home']
```
